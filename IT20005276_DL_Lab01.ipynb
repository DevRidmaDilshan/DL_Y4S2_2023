```python
import numpy as np

random_array = np.random.exponential(scale=1.0, size=(4, 4))
```

```python
import matplotlib.pyplot as plt

random_array_100k = np.random.exponential(scale=1.0, size=100000)

plt.hist(random_array_100k, density=True, bins=100, histtype="step", color="blue", label="exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="orange", label="uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="green", label="normal")
plt.axis([0, 10, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("3D Plot of Z = X^2 + Y^2")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Create a dictionary with your data
pokemon_data = {
    'Name': ['Pikachu', 'Charmander', 'Squirtle', 'Bulbasaur', 'Jigglypuff'],
    'HP': [35, 39, 44, 45, 115],
    'Attack': [55, 52, 48, 49, 45],
    'Defense': [40, 43, 65, 49, 20],
    'Sp. Atk': [50, 60, 50, 65, 45],
    'Sp. Def': [50, 50, 64, 65, 25],
    'Speed': [90, 65, 43, 45, 20]
}

# Create a DataFrame from the dictionary
data = pd.DataFrame(pokemon_data)

# Assuming you have the necessary data loaded into a DataFrame called 'data'
correlation_matrix_pearson = data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')
correlation_matrix_spearman = data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.heatmap(correlation_matrix_pearson, annot=True, cmap='coolwarm', center=0)
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(correlation_matrix_spearman, annot=True, cmap='coolwarm', center=0)
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
